#### Yr_Mo_Dy을 판다스에서 인식할 수 있는 datetime64타입으로 변경하라.

아래의 문제 풀이를 위해 임의로 이상치를 설정한다.

```python
wind_df.loc[10:1000, 'Yr'] = wind_df.loc[10:1000, 'Yr'] - 50
```

여기서는 Yr, Mo, Dy를 합쳐 Yr_Mo_Dy란 변수로 날짜 형식을 만든 후 datetime으로 변환한다.

```python
wind_df['Yr'] = '19' + wind_df['Yr'].astype(str)

Yr_Mo_Dy = wind_df[['Yr', 'Mo', 'Dy']].apply(lambda row: '-'.join(row.values.astype(str)), axis=1)

wind_df = wind_df.drop(['Yr', 'Mo', 'Dy'], axis=1)

wind_df.insert(0, 'Yr_Mo_Dy', Yr_Mo_Dy)

wind_df
```

```python
wind_df.Yr_Mo_Dy = pd.to_datetime(wind_df.Yr_Mo_Dy)

wind_df['Yr_Mo_Dy'].head()
```

#### Yr_Mo_Dy에 존재하는 년도의 유일값을 모두 출력하여라.

```python
wind_df.Yr_Mo_Dy.dt.year.unique()
```

#### Yr_Mo_Dy에 년도가 1961년 미만의 경우에는 수집된 데이터의 범주 밖이다. 해당경우의 값은 50을 더해 새롭게 날짜를 Yr_Mo_Dy 컬럼에 정의하라.

```python
def fix_century(x):
    import datetime
    
    year = x.year + 51 if x.year < 1960 else x.year
    return pd.to_datetime(datetime.date(year, x.month, x.day))


wind_df['Yr_Mo_Dy'] = wind_df['Yr_Mo_Dy'].apply(fix_century)

wind_df.head(4)
```

#### 년도별 각컬럼의 평균값을 구하여라.

```python
wind_df.groupby(wind_df.Yr_Mo_Dy.dt.year).mean()
```

#### weekday 컬럼을 만들고 요일별로 매핑하라. ( 월요일 : 0 ~ 일요일 : 6)

```python
wind_df['weekday'] = wind_df.Yr_Mo_Dy.dt.weekday

wind_df['weekday'].head(3).to_frame()
```

#### weekday컬럼을 기준으로 주말이면 1, 평일이면 0 의 값을 가지는 WeekCheck 컬럼을 만들어라.

```python
wind_df['WeekCheck'] = wind_df['weekday'].map(lambda x : 1 if x in [5,6] else 0)

wind_df['WeekCheck'].head(3).to_frame()
```

#### 년도, 일자 상관없이 모든 컬럼의 각 달의 평균을 구하여라.

```python
wind_df.groupby(wind_df.Yr_Mo_Dy.dt.month).mean()
```

#### 모든 결측치는 컬럼기준 직전의 값으로 대체하고 첫번째 행에 결측치가 있을경우 뒤에있는 값으로 대채하라.

```python
wind_df = wind_df.fillna(method='ffill').fillna(method='bfill')

wind_df.info()
```

```python
wind_df.isnull().sum().to_frame()
```

#### 년도 - 월을 기준으로 모든 컬럼의 평균값을 구하여라.

```python
wind_df.groupby(wind_df.Yr_Mo_Dy.dt.to_period('M')).mean()
```

#### RPT 컬럼의 값을 일자별 기준으로 1차 차분하여라.

```python
wind_df['RPT'].diff().to_frame()
```

#### RPT와 VAL의 컬럼을 일주일 간격으로 각각 이동평균한값을 구하여라.

```python
wind_df[['RPT','VAL']].rolling(7).mean()
```
