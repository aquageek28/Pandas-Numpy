### Library & Data Import

```python
housing_df = pd.read_csv('../../../Datasets/CaliforniaHousing.csv')
```

### 데이터프레임 칼럼 생성과 수정

```python
# 신규 칼럼 생성하기 - 기본

housing_df['Age_by_0'] = 0

housing_df
```

```python
# 신규 칼럼 생성하기 - 응용

housing_df['Age_div_by_10'] = (housing_df['housing_median_age'] // 10).astype(int)

housing_df
```

```python
# 컬럼 데이터 수정하기

housing_df['Age_div_by_10'] = housing_df['Age_div_by_10'] * 10

housing_df
```

```python
# 신규 컬럼에 대한 value_counts()

value_counts = housing_df['Age_div_by_10'].value_counts()

value_counts
```

### 데이터 프레임 삭제

```python
# 컬럼 데이터 삭제하기

housing_drop_df = housing_df.drop('Age_by_0', axis=1)

housing_drop_df
```

```python
# 복수 컬럼 데이터 삭제하기
# drop()의 inplace를 True로 바꾸면 원 데이터프레임의 데이터도 같이 삭제, Default = False

drop_result = housing_df.drop(['Age_by_0', 'Age_div_by_10'], axis=1, inplace=True)

drop_result
```

```python
housing_drop_df = housing_df.drop([0, 1, 2], axis=0)

housing_drop_df
```

### 데이터프레임 데이터 정렬과 집계

```python
# 오름차순 정렬

housing_sorted = housing_df.sort_values(by=['housing_median_age'])

housing_sorted
```

```python
# 복수 컬럼 내림차순 정렬

housing_sorted = housing_df.sort_values(by=['housing_median_age', 'total_rooms'], ascending=False)

housing_sorted
```

```python
# 집계함수 적용

housing_df = housing_df.drop('ocean_proximity', axis=1)
new_df = (housing_df.sum() / housing_df.count())

housing_df
```

```python
# 특정 컬럼집계 함수 적용

housing_df[['housing_median_age', 'total_rooms']].mean()
```

```python
# groupby() 함수 적용

housing_groupby = housing_df.groupby('housing_median_age').mean()

housing_groupby
```

```python
# 특정 컬럼 groupby() 함수 적용

housing_groupby = housing_df.groupby('housing_median_age')[['total_rooms', 'total_bedrooms']].mean()

housing_groupby
```

```python
# 특정 컬럼 groupby(), agg() 함수 적용

housing_groupby = housing_df.groupby('housing_median_age')['total_rooms'].agg([min, max, sum])

housing_groupby
```

### 데이터프레임 결측치 처리하기

```python
# 결측치 여부 확인

housing_df['Age_na'] = np.nan
housing_df.isna()
```

```python
# 결측치 개수 확인

housing_df.isna().sum()
```

```python
# 결측치 데이터 대체하기

housing_df['Age_na'] = housing_df['Age_na'].fillna(housing_df['housing_median_age'].mean())

housing_df
```
