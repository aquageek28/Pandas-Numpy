### Library & Data Import

```python
import pandas as pd
```

```python
df = pd.read_csv('../../Datasets/OlympicMedals.csv')
```

```python
df
```

#### 데이터에서 한국 KOR 데이터만 추출하라.

```python
kr = df[df.Country=='KOR']

kr
```

#### 한국 올림픽 메달리스트 데이터에서 년도에 따른 medal 갯수를 데이터프레임화 하라.

```python
new_df = kr.pivot_table(index='Year',columns='Medal',aggfunc='size').fillna(0)

new_df
```

#### 전체 데이터에서 sport종류에 따른 성별수를 구하여라.

```python
new_df = df.pivot_table(index='Sport',columns='Gender',aggfunc='size')

new_df
```

#### 전체 데이터에서 Discipline종류에 따른 따른 Medal수를 구하여라.

```python
new_df = df.pivot_table(index='Discipline',columns='Medal',aggfunc='size')

new_df
```
