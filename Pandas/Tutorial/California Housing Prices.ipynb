#### 데이터프레임 칼럼 생성과 수정

```python
# 신규 칼럼 생성하기 - 기본
df['Age_by_0'] = 0

df
```

```python
# 신규 칼럼 생성하기 - 응용
df['Age_div_by_10'] = (df['housing_median_age'] // 10).astype(int)

df
```

```python
# 컬럼 데이터 수정하기
df['Age_div_by_10'] = df['Age_div_by_10'] * 10

df
```

```python
# 신규 컬럼에 대한 value_counts()

value_counts = df['Age_div_by_10'].value_counts()

value_counts
```
