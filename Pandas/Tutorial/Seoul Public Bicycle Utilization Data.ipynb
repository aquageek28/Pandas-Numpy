### Library & Data Import

```python
import pandas as pd
import numpy as np
```

```python
bike_df = pd.read_csv('../../Datasets/Bike.csv')

bike_df
```

```python
bike_df.info()
```

####대여일자별 데이터의 수를 데이터프레임으로 출력하고, 가장 많은 데이터가 있는 날짜를 출력하라.

```python
bike_df.loc[:,['대여일자', '대여시간']].groupby(by=['대여일자']).count()
```

#### 각 일자의 요일을 표기하고 ( ‘Monday’ ~’Sunday’ )  ‘day_name’ 컬럼을 추가하고 이를 이용하여 각 요일별 이용 횟수의 총합을 데이터 프레임으로 출력하라.

```python
bike_df['대여일자'] = pd.to_datetime(bike_df['대여일자'])
bike_df['day_name']  = bike_df['대여일자'].dt.day_name()

bike_df
```

```python
result =  bike_df.day_name.value_counts().to_frame()

result
```

#### 각 요일별 가장 많이 이용한 대여소의 이용횟수와 대여소 번호를 데이터 프레임으로 출력하라.

```python
new_df = bike_df.groupby(['day_name','대여소번호']).size().to_frame('size').sort_values(['day_name','size'],ascending=False).reset_index()

new_df
```

#### 6월 7일 ~10대의 “이용건수”의 중앙값은?

```python
new_df = bike_df[['대여일자', '연령대코드', '이용건수']]

new_df
```

```python
result = new_df[(new_df['대여일자'] == '2021-06-07') & (new_df['연령대코드'] == '~10대')].이용건수.median()
```

```python
result
```

#### 평일 (월~금) 출근 시간대(오전 6,7,8시)의 대여소별 이용 횟수를 구해서 데이터 프레임 형태로 표현한 후 <br> 각 대여시간별 이용 횟수의 상위 3개 대여소와 이용횟수를 출력하라.

```python
new_df = bike_df[(bike_df.day_name.isin(['Tuesday', 'Wednesday', 'Thursday', 'Friday','Monday'])) & (bike_df.대여시간.isin([6,7,8]))]
result = new_df.groupby(['대여시간','대여소번호']).size().to_frame('이용 횟수')

answer = result.sort_values(['대여시간','이용 횟수'],ascending=False).groupby('대여시간').head(3)

answer
```

#### 이동거리의 평균 이상의 이동거리 값을 가지는 데이터를 추출하여 추출데이터의 이동거리의 표본표준편차 값을 구하여라.

```python
dist_mean = bike_df['이동거리'].mean()

result = bike_df[bike_df['이동거리'] >= dist_mean].이동거리.std()

result
```

#### 남성(‘M’ or ‘m’)과 여성(‘F’ or ‘f’)의 이동거리값의 평균값을 구하여라

```python
bike_df['성별'] = bike_df['성별'].str.upper()

bike_df.loc[bike_df['성별'] == 'M', '성별'] = '남'
bike_df.loc[bike_df['성별'] == 'F', '성별'] = '여'

bike_df.groupby(by=['성별']).agg({'이동거리' : np.mean})
```
