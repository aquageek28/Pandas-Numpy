### Library & Data Import

```python
import pandas as pd
```

```python
body_df = pd.read_csv('../../Datasets/Body.csv')

body_df
```

#### 전체데이터의 수축기혈압(최고) - 이완기혈압(최저)의 평균을 구하여라.

```python
body_df.info()
```

```python
diff = body_df['수축기혈압(최고) : mmHg'] - body_df['이완기혈압(최저) : mmHg']

result = diff.mean()

print(result)
```

#### 50~59세의 신장평균을 구하여라.

```python
body_df[(body_df['측정나이'] >=  50) & (body_df['측정나이'] <= 59)]['신장 : cm'].mean()
```

#### 연령대 (20~29 : 20대 …) 별 인원수를 구하여라.

```python
body_df['연령대']  = body_df['측정나이'] // 10 * 10
result = body_df['연령대'].value_counts()

print(result)
```

#### 연령대 (20~29 : 20대 …) 별 등급의 숫자를 데이터 프레임으로 표현하라.

```python
body_df.groupby(by=['연령대', '등급']).size().to_frame('size')
```

#### 남성 중 A등급과 D등급의 체지방률 평균의 차이(큰 값에서 작은 값의 차)를 구하여라.

```python
male = body_df['측정회원성별'] == 'M'
a_rank = body_df['등급'] == 'A'
d_rank = body_df['등급'] == 'D'

diff = (body_df[male & a_rank]['체지방율 : %'].mean()) - (body_df[man & d_rank]['체지방율 : %'].mean())

result = print(abs(diff))
```

#### 여성 중 A등급과 D등급의 체중의 평균의 차이(큰 값에서 작은 값의 차)를 구하여라.

```python
female = body_df['측정회원성별'] == 'F'
a_rank = body_df['등급'] == 'A'
d_rank = body_df['등급'] == 'D'

diff = (body_df[female & a_rank]['체중 : kg'].mean()) - (body_df[female & d_rank]['체중 : kg'].mean())

result = print(abs(diff))
```

#### BMI 는 자신의 몸무게(kg)를 키의 제곱(m)으로 나눈값이다. <br>데이터의 BMI 를 구한 새로운 컬럼을 만들고 남성의 BMI 평균을 구하여라.

```python
body_df['BMI'] = body_df['체중 : kg'] / (body_df['신장 : cm']/100)**2

body_df
```

```python
result = body_df[(body_df['측정회원성별'] == 'M')]['BMI'].mean()

print(result)
```

#### BMI 보다 체지방율이 높은 사람들의 체중평균을 구하여라.

```python
result = body_df[body_df['BMI'] < body_df['체지방율 : %']]['체중 : kg'].mean()

print(result)
```

#### 남성과 여성의 악력 평균의 차이를 구하여라.

```python
grip_m = body_df[(body_df['측정회원성별'] == 'M')]['악력D : kg'].mean()
grip_w = body_df[(body_df['측정회원성별'] == 'F')]['악력D : kg'].mean()

result = grip_m - grip_w

print(result)
```

#### 남성과 여성의 교차윗몸일으키기 횟수의 평균의 차이를 구하여라.

```python
wkup_m = body_df[(body_df['측정회원성별'] == 'M')]['교차윗몸일으키기 : 회'].mean()
wkup_w = body_df[(body_df['측정회원성별'] == 'F')]['교차윗몸일으키기 : 회'].mean()

result = wkup_m - wkup_w

print(result)
```
