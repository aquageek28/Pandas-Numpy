#### Iris 데이터 세트에서 species 변수 값별로 petal_width 상위 25% 값을 구한 후, 가장 큰 값과 가장 작은 값 사이의 차이를 계산하시오.

```python
Iris_df.groupby('species')['petal_width'].quantile([0.75]).to_frame()
```

```python
ans = int(Iris_df.groupby('species')['petal_width'].quantile([0.75]).max()) 
                    - int(Iris_df.groupby('species')['petal_width'].quantile([0.75]).min())

print(ans)
```

#### Iris 데이터 세트를 sepal_width 변수 값으로 내림차순한 후 위에서부터 100개 행을 학습데이터로 분리하고, <br> 학습데이터에서 sepal_width의 표준편차 값을 구한 후 50을 곱한 값을 계산하시오. (최종 계산 값의 소수점 이하는 반올림한다.)

```python
sorted_data = Iris_df.sort_values(by='sepal_width', ascending=False).reset_index(drop=True)

sorted_data
```

```python
sorted_data_split = sorted_data.loc[0:99] # 또는 sorted_data.head(100)

sorted_data_split
```

```python
ans = int(sorted_data_split['sepal_width'].std() * 50)

print(ans)
```

#### Iris 데이터 세트에서 species 변수 값별로 petal_length 상위 10개 행(큰 값)의 평균값을 구한 후, 평균값의 합을 계산하시오. <br> (최종 계산값의 소수점 이하는 반올림한다.)

```python
Iris_df = Iris_df.sort_values(by=['species', 'petal_length'], ascending = [True, False])

Iris_df
```

```python
Iris_df_top10 = Iris_df.groupby('species').head(10)

Iris_df_top10
```

```python
print(round(Iris_df_top10.groupby('species')['petal_length'].mean().sum()))
```

#### Iris 데이터 세트에서 sepal_length 변수가 이상치를 가지는 데이터 행 수를 계산하시오. <br> (단, 이상치 범위는 (평균 + 2*표준편차)이상이거나 (평균 - 2*표준편차)이하이다.)

```python
mean = Iris_df['sepal_length'].mean()
std = Iris_df['sepal_length'].std()

crit1 = mean + 2 * std
crit2 = mean - 2 * std

ans = Iris_df[(Iris_df['sepal_length'] >= crit1) | (Iris_df['sepal_length'] <= crit2)]

ans
```
