#### Wine 데이터 세트에서 total_phenols 결측치를 제거하고 alcalinity_of_ash 변수의 결측치는 중앙값으로 대체한 후, <br> alcalinity_of_ash 평균값을 계산하시오.(최종 계산 값의 소수점 이하는 버린다.)

```python
data_drop = wine_df[wine_df['total_phenols'].notnull()]

med = data_drop['alcalinity_of_ash'].median()

print(med)
```

```python
data_drop.loc[data_drop['alcalinity_of_ash'].isnull(), 'alcalinity_of_ash'] = med

ans = round(data_drop['alcalinity_of_ash'].mean(), 3)

print(ans)
```

#### Wine 데이터 세트에서 alcohol 값이 alcohol 평균 값 이상이고, <br>color_intensity 값이 color_intensity 평균 값 이상인 데이터만 남겼을 때, 가장 많은 class 값을 구하시오.

```python
mean_val1 = wine_df['alcohol'].mean()
mean_val2 = wine_df['color_intensity'].mean()

print(mean_val1, mean_val2)
```

```python
data_filtered = wine_df[(wine_df['alcohol'] >= mean_val1) & (wine_df['color_intensity'] >= mean_val2)]

data_filtered
```

```python
data_filtered['class'].value_counts()
```

```python
ans = data_filtered['class'].value_counts().idxmax() # idxmax() 인덱스의 최대값을 구하는 함수

print(ans)
```

#### Wine 데이터 세트에서 class 1에 해당하는 데이터의 ash 평균값과 Wine 데이터 세트의 모든 결측치를 제거한 후의 <br> ash 값의 평균값의 차이를 절댓값을 계산하시오. (최종 계산 값의 소수점 아래 4번째 자리에서 반올림한다.)

```python
data_filtered.head(1).T
```

```python
val1 = wine_df[wine_df['class'] == 1]['ash'].mean()
val2 = wine_df.dropna()['ash'].mean()

print(val1, val2)
```

```python
print(round(abs(val1-val2), 3))
```

#### Wine 데이터 세트에서 alcohol 변수의 상위에서 20번째 값 alcohol을 오름차순 정렬했을 때 20번째에 위치한 값)으로 상위 20개의 값을 변환한 후, proanthocyanins가 2 이하인 데이터를 추출하여 alcohol의 평균값을 계산하시오. (최종 계산 값의 소수점 아래 3번째 자리에서 반올림 한다.)

```python
data_sorted = wine_df.sort_values(by='alcohol', axis=0, ascending = True)

data_sorted
```

```python
data_sorted = data_sorted.reset_index(drop=True)

data_sorted.head(10)
```

```python
data_sorted.loc[0:19, 'alcohol'] = data_sorted.loc[19, 'alcohol']

data_sorted.head(10)
```

```python
ans = round(data_sorted[data_sorted['proanthocyanins']<=2]['alcohol'].mean(), 2)

print(ans)
```

#### Wine 데이터 세트에서 class 1에 해당하는 데이터의 proline 변수의 제 3사분위수와 제 1사분위수 차를 구하고, <br> class 2에 해당하는 데이터의 proline 변수의 제 3사분위수와 제 1사분위 수 차를 구하여 두 값의 차이를 절댓값을 계산하시오.

```python
class1_q1, class1_q3 = wine_df[wine_df['class'] == 1]['proline'].quantile([0.25, 0.75])
class2_q1, class2_q3 = wine_df[wine_df['class'] == 2]['proline'].quantile([0.25, 0.75])

ans = abs((class1_q3 - class1_q1) - (class2_q3 - class2_q1))

print(ans)
```
